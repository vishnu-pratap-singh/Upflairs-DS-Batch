### Create DataFrame 
** display first 2 row 
** Display only Name and Marks columns  
** Add a new column named “Result”  
** Display students with marks greater than 80  

```python
import pandas as pd

# 1. Create the DataFrame using pd.DataFrame()
data = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sneha'],
    'Age': [20, 21, 19, 22],
    'Marks': [85, 90, 78, 88]
}
df = pd.DataFrame(data)
print("--- Original DataFrame ---")
print(df, "\n")

# Task 1: Display first 2 rows using head()
print("--- Task 1: First 2 Rows ---")
print(df.head(2), "\n")

# Task 2: Display only Name and Marks columns (Column Selection)
print("--- Task 2: Name and Marks Columns ---")
print(df[['Name', 'Marks']], "\n")

# Task 3: Add a new column named “Result”
# (Assigning 'Pass' if marks are greater than 80, otherwise 'Fail')
df['Result'] = ['Pass' if score > 80 else 'Fail' for score in df['Marks']]
print("--- Task 3: DataFrame with 'Result' Column ---")
print(df, "\n")

# Task 4: Display students with marks greater than 80 (Conditional Filtering)
print("--- Task 4: Students with Marks > 80 ---")
filtered_df = df[df['Marks'] > 80]
print(filtered_df)
```
